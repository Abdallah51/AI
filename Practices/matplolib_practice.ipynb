```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plotting sine curve
#so i need to detemine start points
x=np.arange(0,3*np.pi,0.1)   # x Axis
y=np.sin(x) # y Axis
plt.plot(x,y)
plt.show()
```

```python
#Plotting two curves sine and cosine curves
x=np.arange(0,3*np.pi,0.1)   # x Axis
y_sin=np.sin(x) # y Axis
y_cos=np.cos(x) # y Axis
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.title('sine and cosine curves')  #Add title for my figure
plt.legend(['Sine','Cosine']) # display the legend to distinguish between different plots
plt.xlabel('x axis')  # display x-axis labels
plt.ylabel('y axis')  # display y-axis labels
plt.show()
```

```python
#Subplotting sine and cosine curves
x=np.arange(0,3*np.pi,0.1)   # x Axis
y_sin=np.sin(x) # y Axis
y_cos=np.cos(x) # y Axis
plt.subplot(2,1,1) #(height of curve, width of curve, plot_number  )
plt.plot(x,y_sin)
plt.title('sine Curve')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cosine Curve')
plt.show()
```
